This notebook simply writes preprocessed screenplays to a folder as .txt files.  Gensim models in particular seem to work much better when streaming the data from a file instead of dealing with RAM-heavy local variables. 

```python
import pandas as pd
import numpy as np

root_path = r'C:\Users\bened\DataScience\ANLP\AT2'

df = pd.read_csv(f'{root_path}\\36118_NLP_Spring\\CSVs\\df_aus_preprocessed2.csv', index_col=0)
df.head()
```

```python
import re
import ast
import unidecode

for index, row in df.iterrows():
    title = row['title']
    script = ast.literal_eval(row['stemmed_scripts'])

    filename = title.lower()
    filename = re.sub(' ', '_', filename)
    filename = re.sub(r'[^\w+]', '', filename)

    file_path = f'{root_path}\\preprocessed_screenplays\\{filename}.txt'

    with open(file_path, 'w', encoding='utf-8') as f:
        for line in script:
            cleaned_line = unidecode.unidecode(line)
            f.write(line + '\n')
```
